```python
# i want male,female,child 
def female_male_sex(passenger):
    Age,Sex = passenger
    if Age < 16:
        return 'child'
    else:
        return Sex
```

```python
data1['person'] = data1[['Age','Sex']].apply(female_male_sex,axis =1)
```
